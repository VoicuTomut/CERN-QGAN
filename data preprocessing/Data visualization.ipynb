```python
import numpy as np
from numpy import save
import matplotlib.pyplot as plt
from sklearn.preprocessing import normalize

from typing import List
```

### Row data

```python
features= ["pt_1","eta_1","phi_1","en_1","pt_2","eta_2","phi_2","en_2"]
```

```python
row_sig=np.load(r"4_features_row_npy\x_sig.npy",allow_pickle=True)
row_bkg=np.load(r"4_features_row_npy\x_bkg.npy",allow_pickle=True)
```

```python
len_sig=len(row_sig)
```

```python
fig, ax = plt.subplots(2,4,figsize=(20, 10)) #figsize=(9, 11)
fig.tight_layout(pad=1.7)
plt.subplots_adjust(bottom=0.05, top=0.9)


for i in range(len(features)):
    
    #bkg
    ax[int(i/4),i%4].hist(row_bkg[:len_sig,i],  alpha=0.5, color='b', label='background',bins=80)
    #sig
    ax[int(i/4),i%4].hist(row_sig[:len_sig,i],  alpha=0.5, color='r', label='signal',bins=80)
    
    
    ax[int(i/4),i%4].title.set_text(features[i])
    
fig.suptitle("",fontweight="bold")
plt.legend(loc='upper right',)
    
plt.show()
```

```python
# MinMax normalize 
```

```python
minmax_sig=normalize(row_sig[:, :], norm='max', axis=0)
minmax_bkg=normalize(row_bkg[:, :], norm='max', axis=0)
save(r"4_features_minmax_npy_unsplit\x_minmax_sig.npy",minmax_sig)
save(r"4_features_minmax_npy_unsplit\x_minmax_bkg.npy",minmax_bkg)
```

```python
fig, ax = plt.subplots(2,4,figsize=(20, 10)) #figsize=(9, 11)
fig.tight_layout(pad=1.7)
plt.subplots_adjust(bottom=0.05, top=0.9)


for i in range(len(features)):
    
    #bkg
    ax[int(i/4),i%4].hist(minmax_bkg[:len_sig,i],  alpha=0.5, color='b', label='background',bins=80)
    #sig
    ax[int(i/4),i%4].hist(minmax_sig[:len_sig,i],  alpha=0.5, color='r', label='signal', bins=80)
    
    
    ax[int(i/4),i%4].title.set_text(features[i])
    
fig.suptitle("",fontweight="bold")
plt.legend(loc='upper right',)
    
plt.show()
```
