```python
fig, ax = plt.subplots(2,4,figsize=(20, 10)) #figsize=(9, 11)
fig.tight_layout(pad=1.7)
plt.subplots_adjust(bottom=0.05, top=0.9)


for i in range(len(features)):
    
    #bkg
    ax[int(i/4),i%4].hist(row_bkg[:,i],  alpha=0.5, color='b', label='background')
    #sig
    ax[int(i/4),i%4].hist(row_sig[:,i],  alpha=0.5, color='r', label='signal')
    
    
    ax[int(i/4),i%4].title.set_text(features[i])
    
fig.suptitle("",fontweight="bold")
plt.legend(loc='upper right',)
    
plt.show()
```
